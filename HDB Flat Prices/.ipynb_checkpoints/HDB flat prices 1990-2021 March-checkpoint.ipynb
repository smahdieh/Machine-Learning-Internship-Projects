## Changing the Dataset

```python
dataset = dataset.drop(columns=['month','street_name', 'address', 'price_psm_yearly', 'Core CPI', 
                          'year_gni', 'flat_model','price cpi_adj', 'price_psm cpi_adj', 'bala lease pct', 
                          'price_psm cpi_lease_adj implied' ])
```

```python
dataset = dataset.drop(columns=['price lease_adj implied', 'price_psm lease_adj implied', 
                                'price cpi_lease_adj implied' ])
```

```python
dataset = dataset.drop(columns=['block', 'town_dummy', 'storey_range'])
```

```python
dataset
```

```python
encoder = LabelEncoder()
for i in range(len(dataset.columns)):

    c = dataset.columns[i]
    if(dataset.dtypes[c] == 'object'):

        dataset[c] = encoder.fit_transform(dataset[c])
```

```python
dataset
```

```python
dataset.describe()
```

## Spliting Data to X, y

```python
X = dataset.drop(columns=['resale_price'])
y = dataset['resale_price']
```

```python
X
```

```python
y
```

```python
print('X shape', X.shape)
print('y shape', y.shape)
```

```python
fig = plt.figure()
fig.set_size_inches(16, 8)
sns.distplot(y)
```

## Scaling

```python
Standard_Scaler = StandardScaler()

X[['year']] = Standard_Scaler.fit_transform(X[['year']])
X[['town']] = Standard_Scaler.fit_transform(X[['town']])
X[['flat_type']] = Standard_Scaler.fit_transform(X[['flat_type']])
X[['storey']] = Standard_Scaler.fit_transform(X[['storey']])
X[['area_sqm']] = Standard_Scaler.fit_transform(X[['area_sqm']])
X[['lease_start']] = Standard_Scaler.fit_transform(X[['lease_start']])
X[['lease_rem']] = Standard_Scaler.fit_transform(X[['lease_rem']])
X[['price_psm']] = Standard_Scaler.fit_transform(X[['price_psm']])
```

## Split Dataset to Train , Test

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=111)
```

```python
print('X train shape', X_train.shape)
print('y train shape', y_train.shape)
```

```python
print('X test shape', X_test.shape)
print('y test shape', y_test.shape)
```

## Model Building (Regression)

```python
models={
        'LinearRegression':LinearRegression(),
        #'PolynomialFeaturesr':PolynomialFeatures(),
        'Ridge':Ridge(),
        'Lasso':Lasso(),
        'ElasticNet':ElasticNet(),
        'KNeighborsRegressor':KNeighborsRegressor(),
        'DecisionTreeRegressor':DecisionTreeRegressor(),
        #'SVR':SVR(),
        }
```

```python
scores = {}
```

```python
for i, model in models.items():

    print("model: ", model)
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    #print(classification_report(y_test , y_pred))
    print("r2 score: ", r2_score(y_test, y_pred))
    print("MAE score: ", mean_absolute_error(y_test, y_pred))
    print("MSE score: ", mean_squared_error(y_test, y_pred))
    
    plt.figure()
    fig.set_size_inches(8, 8)
    plt.scatter(y_test, y_pred, alpha=0.5)
    plt.xlabel('y_true')
    plt.ylabel('y_pred')
    plt.show()
    
    print("-" * 50)
    scores[i] = model.score(X_test,y_test)
```

```python
scores
```

```python
scores = pd.DataFrame(scores, index=['Score']).transpose()
```

```python
scores.sort_values(by = ['Score'], ascending = False, inplace = True)
```

```python
plt.figure(figsize = (16,10))
sns.set(font_scale = 3)
s = sns.barplot(x=scores.Score, y = scores.index, data = scores, palette="husl")
s.bar_label(s.containers[0]);
```
